```python
print("Statistiques descriptives (formaté) :\n")

stats = database_.describe().round(2)

print(stats)
```

```python
# Statistiques descriptives
stats = database_.describe()

# Moyenne
mean_values = stats.loc['mean']

# Médiane (50%)
median_values = stats.loc['50%']

# Comparaison
comparison = abs(mean_values - median_values).sort_values(ascending=False)

print("Différence Moyenne vs Médiane :\n")
print(comparison)
```

# Nuage de points sur la colonne ( src_bytes , dst_bytes ) (Normal vs Anomalie)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(
    database_['src_bytes'],
    database_['dst_bytes'],
    alpha=0.5
)

plt.title("Nuage de points : src_bytes vs dst_bytes")
plt.xlabel("src_bytes")
plt.ylabel("dst_bytes")

plt.show()
```
